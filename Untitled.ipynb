## Newton Rapson method - Test 

In order to test newton rapson method and also coments for this, I will make this test. For thiss, take into account the basic equations
$x_{n+1} = x_{n} - \frac{f(x_{n})}{f^{\prime}(x_{n})} $. In order to prove this, we will fin the roots for $y = 4x^{3} + 5x^{2} - 1$, $y^{\prime} = 12 x^{2} +10x $.

```python
def main_f (num):
    return 4*(num**3) + 5*(num**2) - 1
```

```python
def dev_f(num):
    return 12*(num**2) + 10*(num)
```

```python
# Define the variables that will control the iretarion
maxIter= 20
tolerance = 1e-5
x_0 = 1
div_limit = 1e-10
```

```python
# Main iteration block
for i in range(maxIter):
    y =main_f(x_0)
    y_prime = dev_f(x_0)
    if abs(y_prime) < div_limit:
        print ("Div limit reached")
        break
    else:
        x_1 = x_0 - y/y_prime
        if abs(x_0 - x_1) <= tolerance:
            print ("Root found:", x_1)
            break
        else:
            print ("x_1 in this steep: ", x_1)
            x_0 = x_1
```
